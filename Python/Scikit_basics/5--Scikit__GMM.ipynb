```python
# Using Gaussian Mixture Models to seperate the iris's in discrete layouts
```

```python
%matplotlib inline
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
X_val = iris.drop('species', axis =1)
Y_val = iris['species']
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
```

```python
model.fit(X_val)
```

```python
PCs = model.transform(X_val)
```

```python
iris['PC1'] = PCs[ : , 0]
iris['PC2'] = PCs[ : , 1]
```

```python
sns.lmplot('PC1', 'PC2', hue= 'species', data = iris, fit_reg=False)
```

```python
from sklearn.mixture import GaussianMixture as GMM     # 1. Choose the model class
model = GMM(n_components=3) 
```

```python
model.fit(X_val)
```

```python
y_predicted = model.predict(X_val)
iris['GMM_Y'] = y_predicted
```

```python
sns.lmplot('PC1', 'PC2', hue= 'species', col = 'GMM_Y' ,data = iris, fit_reg=False)
```
