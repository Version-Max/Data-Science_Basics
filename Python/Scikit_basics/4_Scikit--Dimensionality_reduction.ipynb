```python
# Here, we will use PCA to reduce the dimensions of the Iris dataset,
# so we can visualize it in 2D without losing much of its originality
```

```python
%matplotlib inline
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
import numpy as np

iris = sns.load_dataset('iris')

X_var = iris.drop('species', axis = 1)
Y_var = iris['species']
```

```python
# Import PCA library
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
```

```python
model.fit(X_var)
X_components = model.transform(X_var)
```

```python
iris['PCA1'] = X_components[: , 0]
iris['PCA2'] = X_components[: , 1]
```

```python
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```
