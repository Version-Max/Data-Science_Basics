Interpolate is used to fill NaN values in DF or Series.

Uses various linear methods

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Fortune_10_8.csv')
df
```

```python
df.interpolate() # Interpolate is only used for numbers
```

```python
# ^ It tried its best to replace the NaN values in ID
```

```python
df.interpolate(method = 'linear')
```

```python
df.interpolate(method = 'time') # This works only on 
# columns that has DateTime type values in the columns
## method='time' works based on the time column
```

```python
type(df.Date[0])
```

```python
df.interpolate(type = 'linear') 
# Linear guesses the values from the same column
```

```python
df
```

```python
df = pd.read_csv('Fortune_10_8.csv', index_col= 'Profit')
df.interpolate(method = 'index') 
# we set the index to 'Profit' column and then set the interpolate
# to guess based on the 'Profit' column
```

```python
df
```

```python
df.interpolate(method = 'nearest')
```

```python
df.interpolate(method = 'polynomial', order=1)
```
