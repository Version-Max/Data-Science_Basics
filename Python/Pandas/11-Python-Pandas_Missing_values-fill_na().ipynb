# fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv('Fortune_5.csv')
df
```

```python
# fillna replaces the NaN values:

df.dropna()
```

```python
df.dropna(axis = 0)
```

```python
df.dropna(axis = 1)
```

```python
# DF.fillna() can replace NaN with a number or a string
```

```python
df.fillna(950)
```

```python
df
```

```python
df.fillna('Not a Number')
```

```python
df.fillna({'Name' : 'REPLACED',
          'Values' : 27})
```

```python
df.fillna(method = 'ffill') # Takes previous value and replaces NaN
```

```python
df
```

```python
df.fillna(method= 'bfill')
```

```python
df.fillna(axis = 1, method= 'ffill') # Fills values with previous column values
```

```python
df
```

```python
# Limit parameter
```

```python
# Limit defines how many NaN values can be replaced in a col
df.fillna(95, limit = 2)
```

```python
df.fillna(5, inplace = True) # Permanent change
df
```
