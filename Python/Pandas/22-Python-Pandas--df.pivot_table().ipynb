Multi-index objects can be stored using pivot_table()

```python
import pandas as pd
```

```python
df = pd.read_csv('Movie_ratings.csv')
df
```

```python
# Applying pivot operation

df.pivot_table(index = 'Genre')
# Aggregates all the columns to a specific Genre
```

```python
df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
# 'count' counts the number of occurances, default is 'mean'
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum')
```

```python
# fill_value is used to replace missing value
df.pivot_table(index = 'Genre', columns = 'Year', fill_value= -5)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value= -555, margins = True)
# 'All' column and row gives us the mean, since mean is the default 'aggfunc'
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value= -5, margins = True, aggfunc = 's')
# 'All' column and row gives us the mean, since sum is the'aggfunc'
```
