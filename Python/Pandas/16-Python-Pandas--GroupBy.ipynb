```python
import pandas as pd
```

```python
df = pd.read_csv('student_results.csv')
df
```

```python
grp1 = df.groupby(by = 'Class')
grp1.groups # This forms a group by the Class 10 | 11 | 12
```

```python
grp2 = df.groupby(by = ['Class', 'Sleeping hrs'])
grp2.groups # This forms a group by the Class 10 | 11 | 12 & Sleeping hrs
```

```python
list(grp2)
```

```python
dict(list(grp2))
```

```python
gr3 = df.groupby('Class').get_group(10) # Group by only class 10
gr3
```
