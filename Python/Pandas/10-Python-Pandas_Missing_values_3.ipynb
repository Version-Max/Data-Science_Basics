```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Fortune_5.csv')
df
```

```python
# Here, we have a few NaNs
df.isnull().sum()
```

```python
# drop_na helps with removing the particular column or row
# having NaN values

df.dropna(axis = 1)
```

```python
df
```

```python
# 0 is row in axis
# 1 is column in axis

df.dropna(axis = 0)
```

```python
df.dropna(axis = 1)
```


We make use of 'how = '
By default 'how = any' <- drops any row or column with any NaN values
By default 'how = sll' <- drops row or column with all NaN values only


```python
df.dropna(how = 'all')
```

```python
df.dropna(how = 'any')
```

```python
df.dropna(axis = 1, how = 'all')
```

```python
# Threshold parameters helps us establish what is the 
# minimum number of Not-Null values we must have
```

```python
df.dropna(thresh = 4)
```

```python
# With subset you can only delete a particular
# column that has a NaN value

df.dropna(subset = ['Name'])
```

```python
df.dropna(subset = ['Profit'])
```

```python
# inplace = True
# is another operation 

df.dropna(axis = 1, inplace= True) # Important this updates the df itself, so changes are permanent
df
```

```python
df.dropna(inplace= True) # Important this updates the df itself, so changes are permanent
df
```
