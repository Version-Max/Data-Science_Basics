```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    'ID' : [1, 2, 3, 4],
    'Name' : ['A', 'B', 'C', 'D'],
    'Class' : [5, 6, 7, 8]
})

df2 = pd.DataFrame({
    'ID' : [3, 4],
    'Name' : ['C', 'D'],
    'Class' : [7, 8]
})
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
# Print only common values from concat

df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
# Prints on the basis of index 0 and 1
```

```python
# join_axes
```

```python
pd.concat([df1, df2], axis = 1, join_axes = [df1.index])
```

```python
pd.concat([df1, df2], axis = 1, join_axes = [df2.index])
```
