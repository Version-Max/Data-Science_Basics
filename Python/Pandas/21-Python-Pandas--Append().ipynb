Used to append rows of another DF to another DF, which returns a new DF object.

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'A' : [1, 2, 3],
    'B' : [10, 20, 30]
})

df2 = pd.DataFrame({
    'A' : [4, 5, 6],
    'B' : [40, 50, 60]
})

display(df1, df2)
```

```python
df1.append(df2)
```

```python
# Make the indexes appear better:
df1.append(df2, ignore_index=True)
```

```python
df2.append(df1, ignore_index=True)
```

```python
# New column names:
df1 = pd.DataFrame({
    'A' : [1, 2, 3],
    'B' : [10, 20, 30]
})

df2 = pd.DataFrame({
    'C' : [4, 5, 6],
    'D' : [40, 50, 60]
})

df1.append(df2, ignore_index=True, sort=False)
```
