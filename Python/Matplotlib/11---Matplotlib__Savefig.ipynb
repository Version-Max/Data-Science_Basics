# How to save figure (plot and charts)?

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import style
```

```python
plt.pie([75, 6, 2, 17])

# Save pie chart
plt.savefig("pie_char",
            dpi = 100,
           facecolor = 'orange') # <- This saves the image
plt.show()
```

```python
# Saving a subplot
```

```python
plt.figure(figsize = (16, 9))

plt.subplot(3, 2, 1)
plt.pie([1, 3, 4])

plt.subplot(3, 2, 2)
plt.pie([3, 4])

plt.subplot(3, 2, 3)
plt.pie([1, 4, 3, 4])

plt.subplot(3, 2, 4)
plt.pie([1, 8, 3, 4])

plt.subplot(3, 2, 5)
plt.pie([ 3, 4])

plt.subplot(3, 2, 6, projection = 'polar', facecolor = 'black')

# Save all the subplots:
plt.savefig("pie_char.jpg", dpi = 150, facecolor = 'orange')

plt.show()
```
