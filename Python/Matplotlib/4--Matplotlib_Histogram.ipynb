```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18, 45, (100))
py_student_age = np.random.randint(15, 40, (100))
```

```python
print(ml_student_age, '\n\n')
print(py_student_age)
```

```python
# Drawing histogram of the above:
plt.hist(ml_student_age)
plt.title("ML Student Histograms")
plt.xlabel("Age ranges")
plt.ylabel("Number of students per age")
```

```python
# Styling the ^ histogram further:
```

```python
# Bins and width:
plt.hist(ml_student_age, bins = [15, 20, 25, 30, 35, 40, 45],
        rwidth = 0.8,
        histtype= 'bar')
plt.title("ML Student Histograms")
plt.xlabel("Age ranges")
plt.ylabel("Number of students per age")
```

```python
# Plot histogram horizaontally -- orientation = 'horizontal'

plt.hist(ml_student_age, bins = [15, 20, 25, 30, 35, 40, 45],
        rwidth = 0.8,
        histtype= 'bar',
        orientation= 'horizontal',
        color = 'orange',
        label = 'ML_Student')
plt.legend() # Used to show the legend
plt.title("ML Student Histograms")
plt.xlabel("Age ranges")
plt.ylabel("Number of students per age")
```

```python
# Plotting both:
# - ML_Student
# - Py_student

plt.hist(py_student_age, bins = [15, 20, 25, 30, 35, 40],
        rwidth = 0.8,
        histtype= 'bar',
        color = 'red',
        label = 'Py_Student')

plt.hist(ml_student_age, bins = [15, 20, 25, 30, 35, 40, 45],
        rwidth = 0.8,
        histtype= 'bar',
        color = 'orange',
        label = 'ML_Student')
plt.legend() # Used to show the legend
plt.title("ML Student Histograms")
plt.xlabel("Age ranges")
plt.ylabel("Number of students per age")
```

```python
# We can plot the above using lists in one plt.hist and side-by-side

# Increase the length of the histogram plots:
plt.figure(figsize=(25, 16))

# Apply grid
from matplotlib import style
#style.use('ggplot')
plt.grid(color = 'white', linestyle = '-', linewidth = 3)


plt.hist([py_student_age, ml_student_age],
        bins = [15, 20, 25, 30, 35, 40, 45],
        rwidth = 0.8,
        histtype= 'bar',
        color = ['red', 'orange'],
        label = ['Py_Student', 'ML_Student'])

plt.legend()
```
