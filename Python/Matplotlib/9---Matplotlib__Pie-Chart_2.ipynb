```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [45, 15, 35, 25, 30]
```

```python
# -A-
# Take out a slice from the pie. Like a pizza slice
# using explode = [SLICE-Distance_from_the_pie]
explode = [0, 0.3, 0, 1, 0]


# -B-
# put custom colors
colors = ['red', 'blue', 'green', 'yellow', 'orange']


plt.pie(class1_students,
        labels = classes,
       explode = explode,
        colors = colors,
        autopct = '%0.1f%%', # Show slice in percentage
        shadow = True, # Show shadow
        radius = 2.5, # Increase radius
        startangle = 45, # What angle should the pie start
        textprops= {'fontsize': 25} # Increase the size of the font
       )
plt.show()
```
