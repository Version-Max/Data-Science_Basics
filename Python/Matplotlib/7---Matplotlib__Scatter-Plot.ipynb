```python
# Break the data and get only 1000 rows
df_google_play_store_apps = pd.read_csv('googleplaystore.csv', nrows = 1000)
df_google_play_store_apps.head()
```

```python
# Note most formats in string.
# So we will take Ratings vs Reviews
```

```python
x = df_google_play_store_apps['Rating']
y = df_google_play_store_apps['Reviews']
```

```python
plt.figure(figsize = (20, 12))

plt.scatter(x, y,
               color = 'red',
               marker = '<',
                s = 200, # 's' is the size of the marker
                alpha = 0.5, # Transparency
                linewidth = 10,
                edgecolors = 'green',
           )

plt.title('Ratings vs Reviews', fontsize = 85)
plt.xlabel('Ratings', fontsize = 45)
plt.ylabel('Reviews', fontsize = 45)
```

```python
# Plotting two other scatter plots

plt.figure(figsize = (16, 9))

plt.scatter(x, y,
               color = 'red',
               marker = '<',
                s = 200, # 's' is the size of the marker
                alpha = 0.5, # Transparency
                linewidth = 10,
                edgecolors = 'green',
           )

z = df_google_play_store_apps['Installs']
plt.scatter(x, z,
               color = 'orange',
               marker = 'o',
                s = 200, # 's' is the size of the marker
                alpha = 0.5, # Transparency
                linewidth = 10,
                edgecolors = 'yellow',
           )

plt.title('Ratings vs Reviews', fontsize = 85)
plt.xlabel('Ratings', fontsize = 45)
plt.ylabel('Reviews & Install', fontsize = 45)
```
