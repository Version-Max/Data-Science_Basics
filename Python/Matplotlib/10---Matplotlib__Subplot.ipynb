```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize = (16, 9))

plt.subplot(3, 2, 1)
plt.pie([1, 3, 4])

plt.subplot(3, 2, 2)
plt.pie([3, 4])

plt.subplot(3, 2, 3)
plt.pie([1, 4, 3, 4])

plt.subplot(3, 2, 4)
plt.pie([1, 8, 3, 4])

plt.subplot(3, 2, 5)
plt.pie([ 3, 4])

plt.subplot(3, 2, 6, projection = 'polar', facecolor = 'black')

plt.show()
```
