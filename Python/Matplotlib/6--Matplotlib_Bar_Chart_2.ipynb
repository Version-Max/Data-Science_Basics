```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
# Change background using: style.use('ggplot')
style.use('ggplot')

# Increase size:
plt.figure(figsize = (16, 9))

plt.bar(classes, class1_students,
       width = 0.8,
       color = 'orange',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 1 students'
       )
```

```python
# Increase size:
plt.figure(figsize = (16, 9))

# Horizontal plot -- Note: You cannot use 'width'
plt.barh(classes, class1_students,
       color = 'orange',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 1 students'
       )
```

```python
# Increase size:
plt.figure(figsize = (16, 9))

plt.bar(classes, class1_students,
       width = 0.8,
       color = 'orange',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 1 students'
       )
plt.legend()
plt.title('Class and its students', fontsize = 80)
plt.xlabel(' Class name ', fontsize = 45)
plt.ylabel(' # of students ', fontsize = 45)
```

```python
# Plot multiple bars now - Class 1, Class 2, Class 3
plt.figure(figsize = (16, 9))

plt.bar(classes, class1_students,
       width = 0.8,
       color = 'orange',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 1 students'
       )

plt.bar(classes, class2_students,
       width = 0.8,
       color = 'blue',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 2 students'
       )

plt.bar(classes, class3_students,
       width = 0.8,
       color = 'yellow',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 5, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 3 students'
       )


plt.legend()
plt.title('Class and its students', fontsize = 80)
plt.xlabel(' Class name ', fontsize = 45)
plt.ylabel(' # of students ', fontsize = 45)
```

```python
# The above overlaps a lot, making the bar plot redundant to read
```

```python
# To plot the 3 Classes side-by-side we make use of their width
# and add the width to the bar
```

```python
total_classes = np.arange(len(classes))
width = 0.2

# Get width for each class to put them side-by-side
class1_width = total_classes
class2_width = total_classes + width 
class3_width = total_classes + width + width
# Note: The above widths should not exceed more than 1.


# Increase size:
plt.figure(figsize=(16, 9))

# ----- Plot the actual classes 
plt.bar(class1_width, class1_students,
       width,
       color = 'yellow',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 2, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 1 students'
       )

plt.bar(class2_width, class2_students,
       width,
       color = 'red',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 2, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 2 students'
       )

plt.bar(class3_width, class3_students,
       width,
       color = 'orange',
       align = 'edge',# Bar starts at the edge of each X-value
       edgecolor = 'black', # Border color
       linewidth = 2, # Border width
       alpha = 1, # Opacity
       linestyle = '-',
       label = 'Class 3 students'
       )
# ----- 

# Give the bars their name using plt.xticks()
plt.xticks(class2_width, classes, fontsize = 20)

plt.legend()
plt.title('Class and its students', fontsize = 50)
plt.xlabel(' Class name ', fontsize = 25)
plt.ylabel(' # of students ', fontsize = 25)
```
